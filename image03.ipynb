```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((512,512,3), np.uint8)
```

```python
cv2.rectangle(image,(100,450),(400,300),(150,150,0),-1)
plt.imshow(image[...,::-1])
```

```python
pt1 = (100,300)
pt2 = (400, 300)
pt3 = (250, 100)
```

```python
triangle_cnt = np.array( [pt1, pt2, pt3] )
```

```python
cv2.drawContours(image, [triangle_cnt], 0, (0,150,150), -1)
plt.imshow(image[...,::-1])
```
